```python
# 삽입 정렬(big(O) = O(n^2))
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
data = [38, 27, 43, 3, 9, 82, 10]
print(insertion_sort(data))  # 출력: [3, 9, 10, 27, 38, 43, 82]
```

```python
# 버블 정렬(big(O) = O(n^2))
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # 마지막 i개 요소는 이미 정렬되어 있으므로 무시
        for j in range(0, n-i-1):
            # 인접한 두 요소를 비교하여 순서가 잘못된 경우 교환
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(data))  # 출력: [11, 12, 22, 25, 34, 64, 90]
```

```python
# 병합 정렬(big(O) = O(n log n))
def merge_sort(a):
    if len(a) > 1:
        mid = len(a) // 2
        g1 = a[:mid]
        g2 = a[mid:]

        merge_sort(g1)
        merge_sort(g2)

        i1 = 0
        i2 = 0
        ia = 0
        while i1 < len(g1) and i2 < len(g2):
            if g1[i1] < g2[i2]:
                a[ia] = g1[i1]
                i1 += 1
            else:
                a[ia] = g2[i2]
                i2 += 1
            ia += 1
        while i1 < len(g1):
            a[ia] = g1[i1]
            i1 += 1
            ia += 1
        while i2 < len(g2):
            a[ia] = g2[i2]
            i2 += 1
            ia += 1
    return a
data = [38, 27, 43, 3, 9, 82, 10]
print(merge_sort(data))  # 출력: [3, 9, 10, 27, 38, 43, 82]
```

```python
# 파이썬 sort() 함수(big(O) = O(n log n))
data = [38, 27, 43, 3, 9, 82, 10]
data.sort()
print(data)  # 출력: [3, 9, 10, 27, 38, 43, 82]
```
