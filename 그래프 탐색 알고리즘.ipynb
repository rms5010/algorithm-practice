```python
def bfs(g, start):
    visited = set()        # 방문한 정점 기록
    queue = [start]        # 큐 역할을 할 리스트

    while queue:
        node = queue.pop(0)  # 리스트 맨 앞에서 꺼내기 → BFS

        if node not in visited:  # 처음 방문한 노드
            print(node, end=' ')
            visited.add(node)

            # 인접 노드 중 아직 방문 안 한 노드를 큐에 추가
            for neighbor in g[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

# 실행 예시
g = {
    1: [2, 3],
    2: [1, 4, 5],
    3: [1],
    4: [2],
    5: [2]
}

bfs(g, 1)
```
