```python
# 딕셔너리를 활용해 학생 번호로 학생 이름 찾기(없는 경우 "?" 반환, 1~100 번호 가정)
def get_name(student_info, target_no):
    if target_no in student_info:
        return student_info[target_no]
    else:
        return "?"

# 테스트
student_info = {
    39:"Justin", 
    14:"John", 
    67:"Mike", 
    105:"Summer"
    }

print(get_name(student_info, 105))  # Summer
print(get_name(student_info, 777))  # ?
```
